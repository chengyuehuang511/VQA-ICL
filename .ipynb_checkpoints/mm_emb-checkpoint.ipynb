```python
with torch.no_grad():
    batch_images = [[test_dataset[0]["image"]], [test_dataset[1]["image"]]]
    batch_text = [eval_model.get_vqa_prompt(question=test_dataset[0]["question"]),
                 eval_model.get_vqa_prompt(question=test_dataset[1]["question"])]
    image_tensor = eval_model._prepare_images(batch_images)
    text_tensor, attention_mask = eval_model._prepare_text(batch_text)

    joint_features = eval_model.__call__(
        lang_x=text_tensor,
        vision_x=image_tensor,
        attention_mask=attention_mask,
        output_hidden_states=True,
    ).hidden_states[-1][:, -1, :].clone()

    print(joint_features.shape)

    joint_features /= joint_features.norm(dim=-1, keepdim=True)
    joint_features = joint_features.cpu().detach().numpy()  # For NCCL-based processed groups, internal tensor representations of objects must be moved to the GPU device before communication takes place.
```
